# Module 4- List of Data Cleaning & Pre-processing Methods

```python
#adjust any data types to fit into the model
#Split data into test and train data
#remove any outliers
```

```python
#replace all columns with "y" or "n" values with 1 or 0
df['Issue_toolset'].replace(['y','n'],[1,0], inplace=True)
df['New_tool_or_old_tool'].replace(['y','n'],[1,0], inplace=True)
df['Single_path_tool'].replace(['y','n'],[1,0], inplace=True)
df
```

```python
#break out data into x variable these are all of the explanatory variables
X=df[['Historical_backlog','Historical_availability','Historical_utilization','Average_process','QTY_products_tool_feeds','Complexity_of_processes','New_tool_or_old_tool','Single_path_tool','Pct_total_wait']]
X
```

```python
#break out data into the y variable, this is the variable we are explaining with the x variables
Y=df[['Issue_toolset']]
Y
```

# Module 5- Data Analysis Method: Prediction Model

```python
#function to select the three best features to use within the model
best_features= SelectKBest(score_func=chi2, k=3)
fit= best_features.fit(X,Y)
```

```python
df_scores= pd.DataFrame(fit.scores_)
df_columns= pd.DataFrame(X.columns)
```

```python
#creates table to show the score of each feature's fit into the model
features_scores= pd.concat([df_columns, df_scores], axis=1)
features_scores.columns= ['Features', 'Score']
features_scores.sort_values(by = 'Score')
```

```python
#update the x variable with just the best three variables
X=df[['Historical_backlog','Historical_availability','Average_process']]
X
```

```python
#splitting the data into testing and training data
X_train, X_test,Y_train, Y_test = train_test_split(X,Y,
                                   random_state=104, 
                                   test_size=0.25)
```

```python
print('X_train : ')
print(X_train.head())
print('')
print('X_test : ')
print(X_test.head())
print('')
print('Y_train : ')
print(Y_train.head())
print('')
print('Y_test : ')
print(Y_test.head())
```

```python
#create and fit model
logreg= LogisticRegression()
logreg.fit(X_train,Y_train.values.ravel())
```

```python
#input testing data into model
y_pred=logreg.predict(X_test)
print (X_test) #test dataset
print (y_pred) #predicted values
```

# Module 6- Generating Reports

```python
#print multiple metrics for model
print('Accuracy: ',metrics.accuracy_score(Y_test, y_pred))
```

```python
#degine and print accuracy
accuracy = metrics.accuracy_score(Y_test, y_pred)
accuracy
```

import numpy as np
from flask import Flask, request, render_template

@app.route('/')
